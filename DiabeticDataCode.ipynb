```python
def missing (orig_diabetic_data):
    missing_number = orig_diabetic_data.isnull().sum().sort_values(ascending=False)
    missing_percent = (orig_diabetic_data.isnull().sum()/orig_diabetic_data.isnull().count()).sort_values(ascending=False)
    missing_values = pd.concat([missing_number, missing_percent], axis=1, keys=['Missing_Number', 'Missing_Percent'])
    return missing_values

missing(orig_diabetic_data)
```
